```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import sys
cur_dir = os.path.dirname(os.path.abspath("__file__"))  # Gets the current notebook directory
src_dir = os.path.join(cur_dir, '../')  # Constructs the path to the 'src' directory
if src_dir not in sys.path:
    sys.path.append(src_dir)
    
from src.PlotState import *
```

```python
def plot_User_AGV_trajectory_scatter(save_path, df):
    for i in range(df['AGV_name'].nunique()):
        agv_num = i+1
        fig, ax = plt.subplots(figsize=(28, 10))

        locals()[f'agv{agv_num}'] = df[df['AGV_name']==f'AGV{agv_num}'].groupby('TimestampID').mean(numeric_only=True).reset_index()

        sns.lineplot(data=locals()[f'agv{agv_num}'], x='AGV_X', y='AGV_Y', ax=ax, sort=False, label='AGV trajectory')
        sns.lineplot(data=locals()[f'agv{agv_num}'], x='User_X', y='User_Y', ax=ax, sort=False, label= 'User trajectory')

        if "Wait time" in locals()[f'agv{agv_num}'].columns:
            cur_df = locals()[f'agv{agv_num}']
            sns.scatterplot(data=cur_df[cur_df['Wait time'] > 0], x='User_X', y='User_Y', ax=ax, s=200, marker="D", linewidth=1, alpha=.5, color=".5", label='Wait state (User)')
            sns.scatterplot(data=cur_df[cur_df['Wait time'] > 0], x='AGV_X', y='AGV_Y', ax=ax, s=100, color=".2", marker="o", label='Wait state (AGV)')
            
        plt.legend()
        
        for k, v in stations.items():
            fig, ax = draw_rectangle(fig, ax, v, 500, 100, k)

        for i, v in enumerate(sidewalks.values()):
            if i == 0:
                fig, ax = add_sidewalk(fig, ax, *v, showlegend=True)
            else:
                fig, ax = add_sidewalk(fig, ax, *v, showlegend=False)

        plt.xlim(0, 15000) 
        plt.ylim(5000, 10000) 

        ax.set_title(f'AGV{agv_num}')
        plt.savefig(os.path.join(save_path, f'AGV{agv_num}.png'))
        # plt.show()
        plt.close()
```

```python
current_directory = os.getcwd()
data_directory = os.path.join(current_directory, "../data", "PandasData", "Modified")
save_directory = os.path.join(current_directory, "../data", "Plots")
if not os.path.isdir(save_directory):
    os.mkdir(save_directory)

for file in os.listdir(data_directory):
    if file.endswith(".pkl"):
        df = pd.read_pickle(os.path.join(data_directory, file))
        fig_path = os.path.join(save_directory, file[:-4])
        try:
            os.mkdir(fig_path)
        except:pass
        plot_User_AGV_trajectory_scatter(fig_path, df)
```
