```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
loss_baseline = np.load('../model/loss_baseline.npy')
loss_state = np.load('../model/loss_state.npy')

len(loss_baseline), len(loss_state)
```

```python

loss_baseline = loss_baseline[:50000]
loss_state = loss_state[:5000]
df = pd.DataFrame({'baseline': loss_baseline}) #, 'state': loss_state
```

```python
fig = plt.figure(figsize=(13, 6))
sns.lineplot(data=df[:50000])
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Loss')
plt.legend(title='Model')
plt.tight_layout()
plt.savefig('../figure/loss0429.png')
plt.show()
```
