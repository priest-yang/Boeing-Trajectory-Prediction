```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
loss_baseline = np.load('../model/loss_baseline.npy')
loss_state = np.load('../model/loss_state.npy')

len(loss_baseline), len(loss_state)
```

```python
loss_baseline[:-1000].mean(), loss_state[:-1000].mean()
```

```python
df = pd.DataFrame({'baseline': loss_baseline, 'state': loss_state})
```

```python
fig = plt.figure(figsize=(13, 6))
sns.lineplot(data=df[:1500])
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Loss')
plt.legend(title='Model')
plt.show()
```
